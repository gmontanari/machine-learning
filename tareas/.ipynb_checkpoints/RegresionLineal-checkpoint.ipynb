```python
import pandas as pd
import numpy as np
import csv
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1.- Cargo los datos

```python
datos = pd.read_csv("regLin.csv")
datos
```

```python
datos.describe()
```

## 2.- Divido los datos en entrenamiento(.75) y pruebas(.25)

```python
X_train, X_test, Y_train, Y_test = train_test_split(datos[["X"]], datos[["y"]], train_size=0.75)
```

```python
Y_test
```

```python
plt.scatter(X_train,Y_train)
```

## 3.- Elijo el modelo de Regresión Lineal y lo entreno

```python
lr = LinearRegression()
modelo = lr.fit(X_train,Y_train)
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_test, lr.predict(X_test), color='red', linewidth=3)
```

```python
lr.intercept_
```

```python
w0 = np.linspace(lr.intercept_[0]-5,lr.intercept_[0]+10,40)
w0
```

```python
error = []
for i in w0:
    y_predecida = lr.coef_[0]*datos['X'] + i
    resta = y_predecida - datos['y']
    cuadrado = resta**2
    error.append(np.mean(cuadrado))
```

```python
plt.plot(w0,error)
plt.scatter(lr.intercept_[0],np.mean(((lr.coef_[0][0]*datos['X'] + lr.intercept_[0] )-datos['y'])**2),color="red")
```
