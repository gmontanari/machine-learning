```python
from random import randint
```

## 1. Crear una lista de valores aleatorios entre 0 y 100

```python
lista1 = [randint(0,100) for i in range(1000)]
lista1
```

## 2. Haga otra lista del mismo tamaño que los categorice como alto bajo medio

```python
categorias=["alto", "medio", "bajo"]  
lista2 = [categorias[randint(0,2)] for i in range(1000)]
lista2
```

## 3. Haga una lista de parejas de valores el primero debe variar entre 0 y 4 y el segundo entre a y d

```python
diccionario= {}
abecedario=["a", "b","c","d"]
  
for i in range(4):
    for j in abecedario:
        diccionario[i,j] = j
        print(diccionario[i,j])
        #diccionario[i]=abecedario[j]
print (diccionario)
```

```python
diccionario
```

## 4. Haga una función que dada una lista regrese la lista invertida

```python
def lista_inversa(lista):
    return lista[::-1]
```

```python
lista_inversa(lista1)
```

```python
lista_inversa(lista2)
```
